```python
import datasets
from IPython.display import display, HTML
from PIL import Image
import io, base64

ds = datasets.Dataset.from_file("data/icon-qa-train.arrow")
print(f"训练集共 {len(ds)} 条样本")
print(f"字段: {ds.column_names}")
```

```python
def img_to_b64(img):
    buf = io.BytesIO()
    img.save(buf, format='PNG')
    return base64.b64encode(buf.getvalue()).decode()

def show_sample(ds, idx):
    s = ds[idx]
    q_b64 = img_to_b64(s['query_image'])
    c0_b64 = img_to_b64(s['choice_image_0'])
    c1_b64 = img_to_b64(s['choice_image_1'])
    html = f"""
    <div style='border:1px solid #ccc; padding:12px; margin:8px 0; border-radius:8px;'>
      <h3>样本 #{idx}</h3>
      <p><b>问题:</b> {s['question']}</p>
      <p><b>选项:</b> {s['choices']}</p>
      <p><b>答案:</b> {s['answer']}</p>
      <table><tr>
        <td style='text-align:center; padding:8px;'>
          <div>题目图片</div>
          <img src='data:image/png;base64,{q_b64}' style='max-height:150px;'/>
        </td>
        <td style='text-align:center; padding:8px;'>
          <div>选项 0</div>
          <img src='data:image/png;base64,{c0_b64}' style='max-height:150px;'/>
        </td>
        <td style='text-align:center; padding:8px;'>
          <div>选项 1</div>
          <img src='data:image/png;base64,{c1_b64}' style='max-height:150px;'/>
        </td>
      </tr></table>
    </div>
    """
    display(HTML(html))
```

```python
# 展示 5 个样本
for i in [0, 10, 50, 100, 200]:
    show_sample(ds, i)
```
